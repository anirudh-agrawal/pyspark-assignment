## Correcting State Names

```python
#Lets map States to their proper name
stateNames= {
    "U.P.":"UTTAR PRADESH",
    "J AND K":"JAMMU AND KASHMIR",
    "W. BENGAL":"WEST BENGAL",
    "DELHI":"NEW DELHI",
    "ARUNACHAL":"ARUNACHAL PRADESH",
    "UTARANCHAL":"UTTARAKHAND",
    "HIMACHAL":"HIMACHAL PRADESH",
    "M.P.":"MADHAYA PRADESH",
    "MAH":"MAHARASHTRA",
    "RAJASTAN":"RAJASTHAN"
}

unionTerritories = {
    "DADRA N. HAVELI" : "DADRA AND NAGAR HAVELI",
    "DAMAN":"DAMAN AND DIU",
    "DIU":"DAMAN AND DIU",
    "PUDUCHERRY":"PUDUCHERRY",
    "CHANDIGARH":"CHANDIGARH",
    "LAKSHADWEEP":"LAKSHADWEEP",
    "BAY ISLANDS":"ANDAMAN AND NICOBAR ISLANDS"
}

andhraDistricts = ["CHITTOR","GUNTUR","ANANTAPUR",
                   "VISAKHAPATNAM","EAST GODAVARI","NELLORE",
                   "CUDDAPAH","PRAKASAM","WEST GODAVARI",
                   "KURNOOL","SRIKAKULAM","KRISHNA","VIZIANAGARAM"]

def mapStateNames(state,district):
    if state=="EX-INDIA" or state=="UNION.T":
        return unionTerritories[district]
    elif district == "LADAKH":
        return "LADAKH"
    elif state == "ANDHRA":
        return "ANDHRA PRADESH" if district in andhraDistricts else "TELANGANA" 
    elif state in stateNames.keys():
        return stateNames[state]
    return state
```

```python
#Create Spark UDF
mapStateNames = F.udf(mapStateNames,T.StringType())
```

```python
#Applying Spark UDF to correct Names
stations = stations.withColumn("STATE NAME",mapStateNames("STATE NAME","DISTRICT NAME"))
```

```python
#Listing out Distinct States
stations.select("STATE NAME").distinct().show(n=40)
```

## Check for errors in District Names

```python
def generateDistrictList(dataframe,state):
    districtlist= (dataframe
                 .filter(F.col("STATE NAME")==state)
                 .select("DISTRICT NAME")
                 .distinct().collect())
    return [row["DISTRICT NAME"] for row in districtlist]

def generateStateMap(dataframe,states):
    districts = {state:generateDistrictList(dataframe,state) for state in states}
    return districts
```

```python
states = [row["STATE NAME"]   for row in stations.select("STATE NAME").distinct().collect()]
```

```python
districts = generateStateMap(stations,states)
```

## Correct District Names

```python
districtNames={
    "GUJARAT"         : {"BHAUNAGAR":"BHAVNAGAR"},
    "MADHAYA PRADESH" :{"HOSHANGBAD":"HOSHANGABAD"},
    "MAHARASHTRA"    : {"BOMBAY %CITY<":"MUMBAI CITY"},
    "ORISSA"         : {"SUNDERGARH":"SUNDARGARH"},
    "TELANGANA"      : {"NIZAMBAD" : "NIZAMABAD"},
    "WEST BENGAL"    : {"24 PARNANAS N" : "24 PARGANAS N",
                        "MANBHUM%PURULI": "PURULIA"}
}
```

```python
def correctDistrictDetails(row):
    
    if row["STATE NAME"] in districtNames:
        if row["DISTRICT NAME"] in districtNames[row["STATE NAME"]]:
            row= row.asDict()
            row["DISTRICT NAME"] = districtNames[row["STATE NAME"]][row["DISTRICT NAME"]]
            return Row(**row)
    if row["STATE NAME"]=="BIHAR" and row["DISTRICT NAME"]=="GODDA":
        row= row.asDict()
        row["STATE NAME"] = "JHARKHAND"
        return Row(**row)
    return row
```

```python
stations = spark.createDataFrame(data = stations.rdd.map(lambda row: correctDistrictDetails(row)))
```

```python
#Checking that all changes have taken place
generateStateMap(stations,states)
```

## Map Weather Stations to closest StationID
Based on geographical coordinates

```python
#Sorting the data on Latitude and Longitude for faster processing
stationId.sort_values(["Latitude","Longitude"],inplace=True)
```

```python
#Generating series of unique latitudes
latitudes = stationId.Latitude.copy()
latitudes = latitudes.drop_duplicates()
len(latitudes)
```

```python
def getClosest(series,value,returnIndex=False):
    """
    Get closest value from a pandas.Series  for the specified `value`
    Set returnIndex = True to get location of returned value in the Series
    """
    n=len(series[series<=value])
    if n==0: 
        return (series.iloc[0],0) if returnIndex else series.iloc[0] 
    if n==len(series): 
        return (series.iloc[-1],n-1) if returnIndex else series.iloc[-1]
    if abs(series.iloc[n]-value)>abs(series.iloc[n-1]-value):
        n-=1
    return (series.iloc[n],n) if returnIndex else series.iloc[n]


def getClosestStation(row):
    """
    Get ID of the closest station for the specified station
    """
    #Getting the row as dictionary
    rowDict = row.asDict()
    
    #get closest latitude
    closestLatitude = getClosest(latitudes,rowDict["Latitude"])
    
    #get partial dataframe for specific latitude
    partDF=stationId[stationId["Latitude"]==closestLatitude]
    
    #get cloest station on that latitude
    closestLongitude,index = getClosest(partDF.Longitude,rowDict["Longitude"],returnIndex=True)
    
    #adding StationId to row
    rowDict["StationId"] = int(partDF["Id"].iloc[index])
    rowDict["Latitude"]  = float(closestLatitude)
    rowDict["Longitude"] = float(closestLongitude)
    
    #rreturn row
    return Row(**rowDict)
```

```python
#Test call
getClosestStation(Row(Longitude=76.895,Latitude=8.505))
```

```python
#apply getClosestStation UDF to underlying RDD of stations DataFrame
weatherStationsWithId = spark.createDataFrame(data= stations.rdd.map(lambda row: getClosestStation(row)))
```
